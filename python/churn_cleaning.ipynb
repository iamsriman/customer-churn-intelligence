```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/telco_churn.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Monthly Charges"]
df["Total Charges"]
df["Tenure Months"]
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace(" ", "_")
df.columns
```

```python
df["Total_Charges"] = pd.to_numeric(df["Total_Charges"], errors="coerce")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Total_Charges"].fillna(df["Total_Charges"].median(), inplace=True)
```

```python
df["Churn_Label"] = df["Churn_Label"].map({"Yes":1,"No":0})
```

```python
df["Churn_Label"].value_counts()
```

```python
df = df.drop(["CustomerID"], axis=1)
```

```python
df.to_csv("../data/clean_churn_data.csv", index=False)
```

```python
df["Churn_Label"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Churn_Label", data=df)

plt.title("Customer Churn Distribution")
plt.show()
```

```python
sns.countplot(x="Contract", hue="Churn_Label", data=df)

plt.title("Churn by Contract Type")
plt.show()
```

```python
sns.histplot(data=df, x="Tenure_Months", hue="Churn_Label", bins=30)

plt.title("Churn vs Tenure")
plt.show()
```

```python
sns.boxplot(x="Churn_Label", y="Monthly_Charges", data=df)

plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)

plt.title("Feature Correlation")
plt.show()
```


1 Month-to-month contracts have highest churn.
2 Customers with tenure < 12 months churn more.
3 High monthly charges correlate with churn.
4 Electronic check payment users churn frequently.
